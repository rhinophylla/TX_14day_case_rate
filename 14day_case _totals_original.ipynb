```python
import pandas as pd
```

The dataset, "Cases over Time by County" is downloaded periodically from https://dshs.texas.gov/coronavirus/additionaldata.aspx .
The downloaded excel file is modfied by removing the header and footer rows as well as the county population column (though this column was not included in recent downloads of the data) as well as all columns containing data prior to 3/15/20.  New column headers are created using a date only.  The row "counties reporting" is deleted and the "Total" row renamed "State of Texas".  The excel file is saved as a csv file. 

```python
# data read into a pandas df
df = pd.read_csv("Texas COVID-19 Case Count Data by County 11_11.csv")
```

```python
# examine the end of the data frame to make sure no blank rows were included
df.tail()
```

```python
# transpose the dataframe 
df_T = df.T
```

```python
# visual check
df_T.tail()
```

```python
# replace the original data frame with the transposed one containing county names as the column index
df = df_T.rename(columns=df_T.iloc[0]).drop(df_T.index[0])
```

```python
# visual check
df.tail()
```

```python
# calculate the new cases each day and store in a new data frame
new_cases_df = df.diff()
```

```python
# visual check 
new_cases_df.tail()
```

```python
# read in the population of each Texas county into a dataframe
pop = pd.read_csv("tx_county_pop.csv")
```

```python
# visual check
pop.head()
```

```python
# transpose the population dataframe
pop_T = pop.T
```

```python
# replace the original population dataframe with the transposed one containing county names as the column index
pop = pop_T.rename(columns=pop_T.iloc[0]).drop(pop_T.index[0])
```

```python
# visual check
pop
```

```python
# visual check
pop.iloc[0, :]
```

```python
# create new dataframe containing the sum of the rolling 14 day average of new cases.  Visualize how many are null values
case_totals = new_cases_df.rolling(14).sum()
case_totals.isnull().sum()
```

```python
# drop columns that contain null values from the case_total dataframe
case_totals.dropna(inplace=True)
```

```python
# check that no more nulls remain 
case_totals.isnull().sum()
```

```python
case_totals.head()
```

```python
# divide each 14 day rolling average sum by the county population.  Store the results in a new dataframe. 
case_county = case_totals.apply(lambda row : row/pop.iloc[0,:], axis=1)
```

```python
# divide each number by 100,000 to create the final metric. Visual check of the results for Denton County. 
case_county_final = case_county * 100000
case_county_final["Denton"]
```

```python
# visual check
case_county_final
```

```python
# export the results to a CSV file to be uploaded into Tableau Public
case_county_final.to_csv("case_county_final_new.csv")
```
